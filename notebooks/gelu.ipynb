```python
from sympy import *
import numpy as np
import seaborn as sns

import math
x = symbols('x')

import matplotlib.pyplot as plt  
```

```python
y = x/2*(1+ erf(x/sqrt(2)))
z=1.12 + x*0
```

```python
y
```

```python
yx = y.diff(x)
```

```python
yx
```

```python
z
```

```python
p1 = plotting.plot(y, show=False)
p2 = plotting.plot(yx, show=False)
p1.append(p2[0])
p1.show()
```

```python
p1,p2,p3,p4 =plot(y,yx, z,-z,(x, -3, 5), ylim=[-3,5],show=False,label='$f(x)$')
#p[1].line_color = 'r'
#p[2].line_color ='g'
#p[3].line_color ='g'
#p.show()
```

```python
x1, y1 = p1.get_points()  
x2, y2 = p2.get_points() 
x3, y3 = p3.get_points() 
x4, y4 = p4.get_points() 
```

```python
sns.set_style('whitegrid')
plt.plot(x1, np.zeros(67),'k')
plt.plot(np.zeros(160), np.arange(-3, 5, 0.05),'k')
plt.plot(x1, y1, '#3F5D8A') 
plt.plot(x2, y2, '#8A3F63') 
plt.plot(x3, 1.12 * np.ones(65), '-.', color='#587A58')  
plt.plot(x4, -1.12 * np.ones(65), '-.', color='#587A58')  
plt.text(1.7, 1.4, r"$max(f'(x)) = 1.12$", fontsize=16, color='#8A3F63')
plt.text(2.2, 3.5, r"$f(x)$", fontsize=18, color='#3F5D8A')
plt.text(2, 0.3, r"$f'(x)$", fontsize=18, color='#8A3F63')
plt.xlim([-3, 5])
plt.ylim([-2, 5])
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('gelu.pdf')
#plt.plot(x4, y4)  
```
